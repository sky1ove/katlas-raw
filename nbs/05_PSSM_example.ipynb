```python
from katlas.data import *
```

```python
df = Data.get_ks_dataset()
```

```python
df['kinase_id'] = df['kinase_uniprot'] + '_' + df['kinase_protein'].str.split().str[0]
```

```python
from katlas.plot import *
```

```python
def plot_k(df,k):
    df_k = df[df.kinase_protein==k].copy()
    pssm = get_prob(df_k,col ='site_seq')
    plot_logo_heatmap(pssm,f"{k}(n={len(df_k)})",figsize=(17-4,10-3))
    save_svg(f"raw/{k}.svg")
    print(f"saved raw/{k}.svg")
```

```python
plot_k(df,'GSK3B')
```
